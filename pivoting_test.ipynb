# Test code for PC3
Demonstration that a linear approximation to a pivoting power-law underestimates the flux at extremes of the energy band, regardless of the direction of the pivoting.

```python
%matplotlib inline
import matplotlib.pyplot as pl
import numpy as np
from scipy.stats import linregress

def powerlaw(x,N,alpha):
    return N*x**-alpha

def linear(x,m,c):
    return m*x+c
```

### Make some power-law points
Note that normalizations are different so that average flux is the same for two powerlaws

```python
x=np.logspace(0,1,100)
y1=powerlaw(xs,10,1)
y2=powerlaw(xs,0.1,-2)

fig=pl.figure(figsize=(10,7))

ax1=pl.subplot(221)
# ax1.set_xscale('log')
ax1.plot(x,y1)

ax2=pl.subplot(222)
# ax2.set_xscale('log')
ax2.plot(x,y2)

pl.show()
```

### Fit power-laws with linear functions
Using scipy.stats.linregress. Most parameters are not of interest, we just want the slope and intercept (m and c)

```python
m1, c1, r1, p1, err1 = linregress(x,y1)
m2, c2, r2, p2, err2 = linregress(x,y2)

Y1=linear(x,m1,c1)
Y2=linear(x,m2,c2)

fig=pl.figure(figsize=(10,7))

ax1=pl.subplot(221)
# ax1.set_xscale('log')
ax1.plot(x,y1)
ax1.plot(x,Y1)

ax2=pl.subplot(222)
# ax2.set_xscale('log')
ax2.plot(x,y2)
ax2.plot(x,Y2)

pl.show()
```

Note that in both cases, the linear fit underestimates the powerlaw at both ends.

### Difference plots

```python

fig=pl.figure(figsize=(10,7))

ax1=pl.subplot(221)
ax1.plot(x,Y1-y1)


ax2=pl.subplot(222)
ax2.plot(x,Y2-y2)

pl.show()
```

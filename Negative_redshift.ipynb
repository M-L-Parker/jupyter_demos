# v/c vs z, to check I'm not insane

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as pl
```

## Define redshift as function of v/c:

```python
def z(beta):
    return ((1+beta)/(1-beta))**0.5-1
```

## Plot results:

```python
betas=np.linspace(-1,0.5,100)

fig=pl.figure(figsize=(10,5))
ax1=pl.subplot(121)
ax1.set_xlabel(r'$\beta$')
ax1.set_ylabel(r'$z$')

ax1.plot(betas,z(betas))
ax1.plot([-1,0.5],[-1,0.5])

ax2=pl.subplot(122)
ax2.set_xlabel(r'$\beta$')
ax2.set_ylabel(r'abs$(z)-$abs$(\beta)$')

ax2.plot(betas,np.abs(z(betas))-np.abs(betas))
pl.show()
```

Left plot shows z as a function of v/c, note that the absolute values of z are less than those of v/c for z<0, and that z->-1 as v->-c. Right plot shows the difference between abs(z) and abs(v/c)
